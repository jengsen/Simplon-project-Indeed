```python
import Levenshtein as lev
import numpy as np
import pandas as pd
from gensim.test.utils import common_texts, get_tmpfile
from gensim.models import Word2Vec
from gensim.utils import tokenize
import os


def leven(mots,dic):    
    trouve_list = set()
    non_trouve_list = set()
    for m in mots:
        trouve = False
        mot_trouve = ""
        for w in dic:
            distance = lev.distance(w, m)
            if distance <= 1:
                trouve = True
                mot_trouve = w   
                break  #s'il trouve un mot il sort de la boucle du dictionnaire
        if trouve :
            tr= open("trouve.txt","a")
            tr.write(m+ os.linesep)  #écrit le mot trouvé et lui met un séparateur de ligne
            tr.close()
            #print("trouve "+ m +"  "+ mot_trouve)
            trouve_list.add(m)
        else:
            ntr= open("nontrouve.txt","a")
            ntr.write(m+ os.linesep)
            ntr.close()
            print("non trouve "+ m)
            non_trouve_list.add(m)    
    return [non_trouve_list, trouve_list]
    

#fichiers de listes de mots (fichiers pour exclure des mots)
tagged = set(pd.read_csv("tagged.txt", header=None)[0])
exclu = set(pd.read_csv("exclu.txt", header=None)[0])
new_dic = set(pd.read_csv('fr_new_dic.txt', header=None)[0])
old_dic = set(pd.read_csv('fr.txt', header=None)[0])

#data
data = pd.read_json('./BDD_14_10_2019.json',lines=True,encoding='utf-8')
df = data.loc[data['Salaire']!='None',:]
df = df.loc[df['Descriptif_du_poste']!="None",:]
#tokens
df["Tokens_description"] = df.apply(lambda x: list(tokenize(x['Descriptif_du_poste'],lowercase=True)), axis=1)

#liste des mots applatie
descriptions = df["Tokens_description"].tolist()  #tableau de lignes
tokens = set([mot for line in descriptions for mot in line]) #prend les mots de chaque ligne et fait une liste de mots

#pipeline d'exclusion de mots
mots_restants = tokens
print("tokens :" + str(len(mots_restants)))
mots_restants = descriptions.difference(tagged)
print("mots_restants - mots taggés :" + str(len(mots_restants)))
mots_restants = mots_restants.difference(exclu)
print("mots_restants - exclu :" + str(len(mots_restants)))
mots_restants = mots_restants.difference(old_dic)
print("mots_restants - old_dic :" + str(len(mots_restants)))
mots_restants = mots_restants.difference(new_dic)
print("mots_restants - new_dic :" + str(len(mots_restants)))

#leven(mots_restants, new_dic)
print("done")
```
